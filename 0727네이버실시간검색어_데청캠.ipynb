```
import requests
from bs4 import BeautifulSoup
```

```
url = 'http://www.naver.com'
r = requests.get(url)
bs = BeautifulSoup(r.text, 'html.parser')
```

```
#li 요소 태그에 class 명이 "realtime_item"인 요소를 lists에 저장
lists = bs.find_all('li', {'class': 'nav_item'}) #속성 지정
```

```
for li in lists:
  #print(li) #html 다 가져옴
  print(li.find('a').text)
```
