```
url = 'https://movie.daum.net/premovie/released'

# ###웹서버에서 접속하는 클라이언트 정보를 학인한다.
# - 브라우저마다 다른 웹 페이지 또는 서비스를 제공하는 사이트가 존재 할 수 있다.
# - 보통 웹은 사용자가 어떤 브라우저나 디바이스를 사용하건 동일한 접근성을 제공하는 것이 원칙.
# - 브라우저와 웹 표준은 서로 완벽하지 않다. 각 기술간의 차이가 존재한다.
# - 서비스 제공 사이트에서 각 브라우저를 감지하는 기능을 사용 할 수 있음.
# - http통신 중 User-Agent 헤더 정보를 이용하여 클라이언트가 어떤 브라우저/디바이스를 사용하는지 알수 있음.
# - 감지된 브라우저는 디바이스 별로 다른 화면으로 서비스 제공
```

```
from fake_useragent import UserAgent
from bs4 import BeautifulSoup
import requests
# ua = {'User-Agent': 
#       'mozilla/5.0 (Windows NT 10.0; Win64; x64)
#       AppleWebKit/537.36 (KHTMLM liek Gecko)'}

# url = 'https://movie.daum.net/premovie/released'

ua = {'user-agent' : UserAgent().ie}
r = requests.get(url, headers=ua)
bs = BeautifulSoup(r.text, 'lxml')
```
