```
'''
크롤링 순서
1. 원하는 웹페이지에 접속 html 데이터를 받아온다
2. 받아온 html을 분석(파싱) 가능한 형태로 가공한다.
3. 원하는 데이터를 추출 한다.
'''
#웹페이지에 접속하기 위한 라이브러리
import urllib.request

#웹섭에 request 하기위한 객체 생성
request = urllib.request.Request('http://www.naver.com')

#request를 통해 urlopen 후 결과를 변수에 저장
response = urllib.request.urlopen(request)


#html_bytes 는 바이트 형태로 저장, 문자열로 사용하기 위해서 utf-8로 디코딩 필요
html_bytes = response.read()
html_str = html_bytes.decode('utf8')

#html_str의 내용을 엔터값(줄바꿈)으로 분리 작업(\n)
html_split = html_str.split('\n')
#리스트성 데이터 


# 분리된 내용을 줄단위로 출력
for line in html_split:
    print(line)

'''
## Requests
- 파이썬에서 웹서버에 접속해서 html을 받을 수 있는 외부 라이브러리
- 파이썬 설치시 설치되는 기본 라이브러리보다 사용성도 편하고 속도도 빠름.
- pip install requests
'''
import requests

response = requests.get('https://www.naver.com/index.htm') #어떤 url을 가져올지가 중요
# 뉴스 : https://www.naver.com/main/list. ~ 

print(response.status_code)

print(response.headers['content-type'])

print(response.encoding)

print(response.text)

'''
status_code
200 : 정상(제일 중요)
404 : index.html이 기본 리소스가 아니라는 걸 알게됨
    
application error code
500대
'''

'''
#BeautifulSoup
 - html을 분석 가능한 형태로 가공, 추출 할 수 있는 외부 라이브러리
 - pip install BeautifulSoup4 or bs4
 - BeautifulSoup 사용시 해석 라이브러리(파서)를 지정 해 줘야 한다.
 - 파서로는 기본적인 html.parse, lxml 등이 있음(lxml 파서는 외부 라이브러리)

정확성 : html5 파서
속도 : lxml
태그지정 : html 파서
'''
import requests
from bs4 import BeautifulSoup

reponse = requests.get('https://www.naver.com')
bs = BeautifulSoup(response.text, 'html.parser')

#a태그만 다 들고 와라
print(bs.select('a')[0])
```
