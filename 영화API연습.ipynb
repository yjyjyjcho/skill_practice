```
import urllib
import urllib.request as urllib2
import json
import pandas as pd

movieIdListURL = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?key=430156241533f1d058c603178cc3ca0e&openStartDt=2017&itemPerPage=40"

movieIdPage = urllib2.urlopen(movieIdListURL)
movieIdData = json.loads(movieIdPage.read()) #파이썬에 있는 딕셔너리 구조 파싱한 것을 data에 저장
movieIdData

movieIDDF = pd.DataFrame()
movieIDDF = movieIDDF.append(
    {"movieCd":"","movieNm":"","movieNmEn":"","oepnDt":"","peopleNm":""
    }, 
    ignore_index=True)
print(movieIdData)
print('\n')
print(movieIdData["movieListResult"]["movieList"])
print('\n')
print(movieIdData["movieListResult"]["movieList"][0]["movieNm"])
print('\n')
num = len(movieIdData["movieListResult"]["movieList"])
print(num)
```

```
list = [
        {'Name':'Tom', 'age':30},
        {'Name':'jack', 'age':20}
]
 # Name이 Tom인 Dictionary를 List에서 가져오기
tom = (item for item in list if item['Name'] == 'Tom')
 # 만약에 값이 없을 경우 False가 dict에 저장된다. ... 
# Tom이라는 값이 더 있을 경우 next시 그 다음 Dictionary를 가져오게 된다. ... 
dict = next(tom, False) 
dict {'Name': 'Tom', 'Age': 30} 
dict['Age'] 30

```

```

for i in range(0,num):
    movieIDDF.iloc[i,"movieCd"]=movieIdData["movieListResult"]["movieList"][i]["movieCd"]
    movieIDDF.iloc[i,"movieNm"]=movieIdData["movieListResult"]["movieList"][i]["movieNm"]
    movieIDDF.iloc[i,"movieNmEn"]=movieIdData["movieListResult"]["movieList"][i]["movieNmEn"]
    movieIDDF.iloc[i,"openDt"]=movieIdData["movieListResult"]["movieList"][i]["openDt"]

    if movieIdData["movieListResult"]["movieList"][i]["directors"]!=[]:
        movieIDDF.iloc[i,"peopleNm"]=movieIdData["movieListResult"]["movieList"][i]["directors"][0]["peopleNm"]
    
    else:
        movieIDDF.iloc[i,"peopleNm"]=""

```
