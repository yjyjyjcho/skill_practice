2-2 dataframe


1. list로 만들기

```
company1 = [['삼성', 2000, '스마트폰'],
            ['현대', 1000, '자동차'],
            ['네이버', 500, '포털']]

pd.DataFrame(company1)
```

```
df1 = pd.DataFrame(company1)
```

```
df1 #데이터프레임 생성후 변수지정을 해줘야 다시 호출 가능함 (시리즈도 마찬가지)
```

제목 컬럼 만들기


```
df1.columns = ['회사명', '매출액', '업종']
df1
```

dict로 만들기

```
company2 = {'기업명':['삼성', '현대', '네이버'],
            '매출액':[2000,1000,500],
            '업종':['스마트폰', '자동차', '포털']}
```

```
df2 = pd.DataFrame(company2)
df2
#딕셔너리형태일 때는 키값이 컬럼명, 데이터프레임도 가로로 선언됨
```
