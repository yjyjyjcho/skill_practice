```
keyword = '빅데이터'
url = 'https://search.naver.com/search.naver?f=&fd=2&filetype=0&nso=so%3Ar%2Ca%3Aall%2Cp%3Aall&query={}&research_url=&sm=tab_nmr&start=1&where=webkr'.format(keyword)
r = requests.get(url)
bs = BeautifulSoup(r.text, 'lxml')

results = []
for li in bs.select('li.sh_web_top'):
    title = li.select('dl > dt > a')[0].text
    title_link = li.select('dl > dt > a')[0]['href']
    summary = li.select('dl > dd.sh_web_passage')[0].text
    results.append([title, title_link, summary])

import pandas as pd
df = pd.DataFrame(results)
df.to_csv('1.csv', header=True, index=False)
```

```
import requests
from bs4 import BeautifulSoup
import pandas as pd

def naver_search(search_word, pages = 1):
    for page in range(pages):
        i =1
        url = 'https://search.naver.com/search.naver?f=&fd=2&filetype=0&nso=so%3Ar%2Ca%3Aall%2Cp%3Aall&query=%EB%B9%85%EB%8D%B0%EC%9D%B4%ED%84%B0&research_url=&sm=tab_pge&start={0}&where=webkr'.format(page)
    url = 'https://search.naver.com/search.naver?where=webkr&sm=tab_jum&query={0}'.format(search_word)
    r = requests.get(url)
    bs = BeautifulSoup(r.text, 'html.parser')
    results = []
    search_naver = bs.select('#elThumbnailResultArea > li')
    for result in search_naver:
        title = result.select('a', {'class':'title_link'})[0].text
        url = result.select('div.web_url > a.txt_url')[0].text
        summary = result.select('dd.sh_web_passage')[0].text
        results.append([title, url, summary])
    column = ['제목', 'URL', '요약']
    df = pd.DataFrame(results, columns = column)
    df.to_excel('네이버검색.xlsx', sheet_name='네이버검색', columns=column, header=True, startrow=1)

    naver_search('빅데이터', 1)
```
