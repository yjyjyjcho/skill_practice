```
# !pip install selenium
from selenium import webdriver
import time
import requests
from bs4 import BeautifulSoup

driver = webdriver.Chrome('C:\selenium\chromedriver.exe') #코렙에선 경로 안먹힘, 자꾸 돌리면 꺼짐 밖으로 뺄것

driver.get('https://www.starbucks.co.kr/store/store_map.do') #중간에 실패하면 이 지점 부터 다시 시작 할것

time.sleep(3)

loca = driver.find_element_by_clas_name('loca_search')
print(loca)
loca.click()
time.sleep(3)

sido = driver.find_element_by_clas_name("sido_area_box")
li = sido.find_element_by_tag_name("li")
print(li[5])
li[5].click()
time.sleep(3)

gugun = driver.find_element_by_class_name('gugun_arae_box')
li = gugun.find_elements_by_tag_name('li')
li[13].click()
time.sleep(3)

html = driver.page_source
bs = BeautifulSoup(html, 'lxml')
entrie = bs.select("ul.quickSearchResultBoxSidoGugun")[0]
li_list = entrie.find_all('li')
for li in li_list:
    print(li.find('p').text)
```
