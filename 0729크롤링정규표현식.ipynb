```
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
```

```
for index in range(1, 2):
    print("https://www.seeko.kr/zboard4/zboard.php?id=mainnews&page=' + str(index) + '&select_arrange=headnum&desc=asc&category=&sn=off&ss=on&sc=off&keyword=&sn1=&divpage=10")
    res = urlopen('https://www.seeko.kr/zboard4/zboard.php?id=mainnews&page=' + str(index) + '&select_arrange=headnum&desc=asc&category=&sn=off&ss=on&sc=off&keyword=&sn1=&divpage=10')

    bs = BeautifulSoup(res, 'html.parser')
    data = bs.find_all('td', 'article_subject')
    for item in data:
      print(re.sub('\[0-9+\]'), ' ' , item.get_text())
#7인치 화면 시대 열려.....
```
