```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


df = pd.read_excel('정보망 강좌정보(2014년~2020년).xls')

df.isnull().sum()

df.sort_values(by=['지역'], axis=0, inplace=True)
```

```
#df : 원본, df_drop : 대분류 and 소분류 결측값 제거 한 것들


df_drop = df.dropna(subset= ['대분류', '소분류'])
df_drop.isnull().sum()

df_drop

df_drop.sort_values(by=['지역'], axis=0, inplace= True)
```
